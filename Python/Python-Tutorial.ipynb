## What is a Program?

 A program is a sequence of instructions (i.e., an algorithm) designed to solve a particular problem in a machine. 
 Examples of such problems include: 

 Controlling the behavior of a particular machine; -- solving a mathematical, scientific or engineering computation problem;
 --solving a sybolic computation problem (searching, compiling, etc).
 When the sequence of instructions is inteded to control a computer (i.e., a particular machine), we refer to the program as a computer program.

## What is computer Programming?

**Computer Programming** Computer Programing is the process of designing a solution to a particular problem by
implementing (i.e., coding) the solution algorithm (i.e., the sequence of instructions) in a target programming 
language so that it can be executed by a computer. Source code is written in one or more programming languages.

**Purpose** The purpose of computer programming is to find a sequence of instruction that will sutomate performing a 
specific task (solve a particular problem) using a computer (broadly defined as any machine/device capable of 
processing instructions, e.g. a processor).

**Software Development** involves programming (which is often referred to as the implementation or 
coding -i.e., writing source code) as well as other activities including formulation of specification (design
inputs definition), algoithm design, verification, testing, validation, quality assurance, maintaining the 
source code, implementation of the built system, etc. Software engineering combines software development methods
with principles and practives of engineering.

## What are the Characteristics of High-Level Languages?

- High-level programming languages allow the programmer to write programs in terms that are more abstract than
assembly language (i.e., higher level of abstraction than assembly).
- The programming synta is closer to natural language or pseudocode. Rather than dealing with hardware-specifics
(registers, memory addresses and callbacks, etc), high-level languages deal with variables, data structures, 
arithmetic and boolean computation, conditional execution, repetition (loops), functions, and objects which are 
hardware independent.
- The focus is usability (the easy with which a person can use the program for the inteded purpose) over optimal
program efficiency.
- Easier to learn and allows programmers to develop applications much more efficiently, with less source code, that
are more portable, and are easier to maintain.
- High-level programming languages incur overhead and often produce less efficient code than assembly. However,
with the modern microprocessor architechutes and compilers, high-level languages frequently produce code 
comparable in efficiency to what most low-level programmers can produce.
- There are still situations that require assembly programming: e/g/. where low-level hardware control is 
necessary, maximum processing speed is critical, and/or there are significant hardware resource constraints 
(e.g., resource limited micro-controllers, etc).

# Part 7: Data Structres: Lists (Arrays), Tuples, Sets, Dictionaries, Strings

Python provides several built-in data structures that allow for efficient storage and manupulation of data. These include:
- **Lists**: Lists are ordered collections of items, which can be of any type. They are defined using square brackets [] and items are separated by commas. Lists are mutable, which means that items can be added, removed, or modified.
- **Tuples**: Tuples are similar to lists but are immutable, meaning the items cannot be modified once they are created. They are defined using parentheses () and items are separated by commas.
- **Sets**: Sets are unordered collections of unique items. They are defined using curly braces {} or the set() function. Sets do not allow duplicate items and are useful for performing mathematical set operations like union and intersection.
- **Dictionaries**: Dictionaries are unordered collections of key-value pairs. They are defined using {} and keys and values are seperated by colons. Dictionaries are extremely useful for storing and retrieving data quickly and are often used to implement data structures like hash tables.
- **Strings**: A string is a sequence of characters. They are defined using either single or double quotes. Python provides a variety of methods for manipulating strings, including string concatenation, slicing, and string formatting.

# Part 9: Error Handling
Error handling is an important concept in programming, allowing you to anticipate and handle potential errors in your code. In Python, errors are represented by exceptions, which are events that triggered when something goes wrong in your code.
There are several ways to handle exceptions in Python, including:
- **Try-except blocks**: You can use a try-except block to enclose a block of code that may rase an exception. If an exception is raised, the code in the except block will be executed.

`try:`
     
   ` # Code that may raise an exception`


`except ExcetionType:`


    # Code to handle the exception

- **Raising Exceptions**: You can raise an exception with the `raise` keyword. This can be useful when you want to indicate that something unexpected has occurred in your code.

`if variable <0:`

    raise ValueError("Variable cannot be negative")

- **The `assert` statement**: You can use the assert statement to check if a certain condition is True. if the condition is False, an AssertionError is raised.

`assert variable >0`

Error handling is an important aspect of programming, and it's especially important in electrical engineering applications where the code may be interacting with external hardware or dealing with large amounts of data. In this chapter, we will disscuss the basics of error handling in Python and provide some examples of how to use it in electrical engineering applications.
There are several types of errors that can occur in Python, including:
- **Sytax Errors**: These occur when the code is not written correctly, and the interpreter cannot understand it.
- **Exceptions**: These occur when something unexpected happens during the execution of the code. The most common way to handle errors in Python is to use a try-except block. the try block contains the code that may cause an error, and the except block contains the code that will be executed if an error occurs.

```python
done = False
while done!= True:
    try:
        x = int(input("Enter a number: "))
        print(x)
        done = True
    except ValueError:
        print("Please enter a number in numerical form without a decimal.")
```

```python
import matplotlib.pyplot as plot
import numpy as np

def projectile_motion(v0, theta0):
    # ProjectileMotion calculates the horizontal range, maximum height,
    # location of the maximum height, and displays a plot of the trajectory
    # for a projectile with the initial velocity v0 and initial angle theta0.
    #
    # Inputs: 
    # v0 = initial velocity
    # theta0 = initial angle
    #
    # Outputs:
    # range = horizontal distance
    # max_height = the maximum y value
    # max_height_x = where on the x axis the max height occurs
    # h_pos = matrix of the x,y positions of the projectile over time.
    
    g=9.8
    theta0 = theta0 * (np.pi/180)
    vx = v0*np.cos(theta0)
    vy = v0*np.sin(theta0)

    # Time of flights
    t_flight = 2*vy/g

    # Calculate Range
    range = vx * t_flight

    # Calculate maximum height
    max_height = vy**2/(2*g)

    # Calculate location of max
    max_height_x = vx*vy/g

    # Calculate the x and y positions over time
    t = np.linspace(0,t_flight,1000)
    x_pos = vx*t
    y_pos = vy*t+(.5*-g*(t**2))

    # Store x and y postions in h_pos matrix
    h_pos = [x_pos,y_pos]

    # Plot the trajectory
    plot.figure(edgecolor='white')
    plot.plot(x_pos,y_pos) 
    plot.xlim([0,25])
    plot.ylim([0,25])
    plot.xlabel("Horizonal Position (m)")
    plot.ylabel("Vertical Position (m)")
    plot.title("Projectile Trajectory")

    return range, max_height,max_height_x,h_pos

```

```python
v0 = 15
theta0=70
range,max_height,max_height_x,h_pos = projectile_motion(v0,theta0)
print(f"Results for the Velocity of {v0} and theta of {theta0} degrees: \nRange: {range:.2f} \nMax Height: {max_height:.2f} \nMax Height Location: {max_height_x:.2f}")
```
